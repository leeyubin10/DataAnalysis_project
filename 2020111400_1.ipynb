```python
import numpy as np
import csv

#  .
f = open('age.csv')
data = csv.reader(f)
next(data)
data = list(data)

mn = 1
result_name = ''
result = 0

for row in data:
    if '2' in row[0] :
        home = np.array(row[106:], dtype=int) / int(row[105].replace(',',''))

for row in data:
    i = 106
    while i < 207 :
        row[i] = int(row[i].replace(',',''))
        i= i+1
    away = np.array(row[106:], dtype=int) / int(row[105].replace(',',''))
    s = np.sum((home-away)**2)
    if s < mn and '2' not in row[0]:
        mn = s
        result_name = row[0]
        result = away

#  
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title('2        ')
plt.plot(home, label='2')
plt.plot(result, label=result_name)
plt.legend()
plt.show()
```
