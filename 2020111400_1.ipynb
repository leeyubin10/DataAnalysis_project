```python
import numpy as np
import csv

# 데이터를 읽어온다.
f = open('age.csv')
data = csv.reader(f)
next(data)
data = list(data)

mn = 1
result_name = ''
result = 0

for row in data:
    if '안양2동' in row[0] :
        home = np.array(row[106:], dtype=int) / int(row[105].replace(',',''))

for row in data:
    i = 106
    while i < 207 :
        row[i] = int(row[i].replace(',',''))
        i= i+1
    away = np.array(row[106:], dtype=int) / int(row[105].replace(',',''))
    s = np.sum((home-away)**2)
    if s < mn and '안양2동' not in row[0]:
        mn = s
        result_name = row[0]
        result = away

# 그래프 시각화
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title('안양2동 지역과 가장 비슷한 여성 인구 구조를 가진 지역')
plt.plot(home, label='안양2동')
plt.plot(result, label=result_name)
plt.legend()
plt.show()
```
